### Extract the columns

```python
df1=df1[['Review Title','Review text','Ratings']]
df2=df2[['Review_Title','Review_Text','Reviewer_Rating']]
df3=df3[['review_title','review_text','reviewer_rating']]
```

### Handle missing values 

```python
df.isna().sum()
```

```python
df['review_title'].fillna("No Title", inplace=True)
df = df.dropna(subset=['review_text'])
df['rating'].fillna(df['rating'].median(), inplace=True)
```

```python
df['review_text'] = df['review_text'].fillna("")
```

### Create the  sentiment column 

```python
df['sentiment'] = df['rating'].apply(lambda x: 1 if x >= 3 else 0)
```

### Clean the text

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
stop_words.remove("not")
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    words = text.split()
    words = [lemmatizer.lemmatize(w) for w in words if w not in stop_words]
    return " ".join(words)
```

```python
df['clean_review'] = df['review_text'].apply(clean_text)
```

```python
df.head()
```

### split the data into  train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( df['clean_review'], df['sentiment'], test_size=0.2, random_state=42)
```

### Vectorization (Text to Vectors)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)
```

### LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

lrmodel = LogisticRegression(class_weight='balanced')
lrmodel.fit(X_train, y_train)
```

### LogisticRegression score 

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix

y_pred = lrmodel.predict(X_test)
lrscore=accuracy_score(y_test, y_pred)
print("Accuracy:",lrscore )
print(classification_report(y_test, y_pred))
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

###  MultinomialNB(naive_bayes)

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
nb.fit(X_train, y_train)

y_pred_nb = nb.predict(X_test)
nb_acc=accuracy_score(y_test, y_pred_nb)
print("Naive Bayes Accuracy:",nb_acc )
print(classification_report(y_test, y_pred_nb))
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred_nb))
```

### LinearSVC

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix

svm = LinearSVC(class_weight='balanced')
svm.fit(X_train, y_train)

y_pred_svm = svm.predict(X_test)
svm_acc=accuracy_score(y_test, y_pred_svm)
print("SVM Accuracy:", svm_acc)
print(classification_report(y_test, y_pred_svm))
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred_svm))
print(confusion_matrix(y_test, y_pred_svm))
```

### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(class_weight='balanced',n_estimators=200, random_state=42)

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
rf_acc= accuracy_score(y_test, y_pred_rf)
print("Random Forest Accuracy:",rf_acc)
print(classification_report(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
```

### KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

y_pred_knn = knn.predict(X_test)
knn_acc=accuracy_score(y_test, y_pred_knn)
print("KNN Accuracy:",knn_acc )
print(classification_report(y_test, y_pred_knn))
print(confusion_matrix(y_test, y_pred_knn))
```

**SVM achieved the highest F1-score and recall for reviews. Therefore it was chosen as the final model.**
